```
import plotly.express as px
import chart_studio.plotly as py
import numpy as np
import pandas as pd
```

```
df3=pd.read_csv("https://raw.githubusercontent.com/napoles-uach/covid19mx/master/estadoslatlonCarranco1.csv")
estados_list=df3["Estado"].to_list()
lat_list=df3["Lat"]
lon_list=df3['Long']
estados_list
```

```
hoy='20200331'
fechas_list=pd.date_range(start='20200301',end=hoy,freq='D').strftime('%Y%m%d').to_list()

fechas_format=pd.date_range(start='20200301',end=hoy,freq='D').strftime('%d'+'/'+'%m'+'/'+'%Y').to_list()
fechas_list

hoy='20200408'
fechas_list2=pd.date_range(start='20200401',end=hoy,freq='D').strftime('%Y%m%d').to_list()

fechas_format2=pd.date_range(start='20200401',end=hoy,freq='D').strftime('%d'+'/'+'%m'+'/'+'%Y').to_list()
fechas_list2
```

```
todos=[]
#list=[]

for i in range(len(fechas_list)):
  df1=pd.read_csv('https://raw.githubusercontent.com/carranco-sga/Mexico-COVID-19/master/Daily%20data/202003/fallecidos_'+fechas_list[i]+'.csv')
  #conteo=[]
  tot_est=[]
  dfmap=pd.DataFrame(df1)
  #dfmap["conteo"]=pd.Series(conteo)
  dfmap=pd.DataFrame(df1)
  dfmap['Lat']=pd.Series(lat_list)
  dfmap['Long']=pd.Series(lon_list)
  dfmap['Fecha']=fechas_format[i]  
  todos.append(dfmap)

for i in range(len(fechas_list2)):
  df1=pd.read_csv('https://raw.githubusercontent.com/carranco-sga/Mexico-COVID-19/master/Daily%20data/202004/fallecidos_'+fechas_list2[i]+'.csv')
  #conteo=[]
  tot_est=[]
  dfmap=pd.DataFrame(df1)
  #dfmap["conteo"]=pd.Series(conteo)
  dfmap=pd.DataFrame(df1)
  dfmap['Lat']=pd.Series(lat_list)
  dfmap['Long']=pd.Series(lon_list)
  dfmap['Fecha']=fechas_format2[i]
  todos.append(dfmap)
todos = pd.concat(todos, ignore_index=True)

#todos = todos.sort_values(by=['Fecha'])
```

```
todos
```

```
fig = px.scatter_geo(todos, lat="Lat",lon="Long", size="Fallecidos",color='Fallecidos',
                     animation_frame="Fecha",scope='north america',range_color=[0,max(todos['Fallecidos'])],
                    # center={'lat':19,'lon':-99},width=800,text = todos['Estado'],
                     projection="natural earth")
#fig.update_layout(width=800, margin={"r":0,"t":0,"l":0,"b":0})
fig.update_layout(title = 'Mapa de Defunciones<br>Fecha: 8/04/2020)')
fig.show()
plot_url = py.plot(fig)
```
